```python
import requests
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup
```

Here, we scrape the voting results by county from Wikipedia

```python
URL = "https://en.wikipedia.org/wiki/2020_United_States_presidential_election_in_California" #URL to be scraped
page = requests.get(URL, headers={'User-Agent': 'Mozilla/5.0'}) #request the URL page using Mozilla useragent
#without user agent, request is blocked returning Error 403

soup = BeautifulSoup(page.content, "html.parser") #use beuatiful soup to parse the html
```

```python
table_found = soup.find_all("table") #find all objects with class "table"
```

```python
table_found[-6] #the relevant table
```

```python
print(table_found[-2].prettify()) #right half of table we are interested in
```

```python
table = pd.read_html(table_found[-6].prettify())[0] #prettify left half of the scraped table into str
table
```

```python
flattened = table.T.reset_index(drop=True).T.iloc[0:,[0,1,3]] #reset index to flatten, and transpose
#choose only percentages for each county
col_names = ["county","dem_pct", "rep_pct"] #set names of new columns
flattened.set_axis(col_names, axis = 1, inplace = True) #assign the new column names
flattened
```

```python
dem_counties = flattened[flattened.dem_pct > flattened.rep_pct].county #narrow down democrat majority counties
list(dem_counties)
```

```python
rep_counties = flattened[flattened.dem_pct < flattened.rep_pct].county #narrow down republican majority counties
list(rep_counties)
```
