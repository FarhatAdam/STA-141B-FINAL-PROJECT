```python
url1 = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/us_state_vaccinations.csv'
df_vac = pd.read_csv(url1)
df_vac
```

```python
df_vac["date"] = df_vac["date"].astype("datetime64")

start_date = "2021-02-01"
end_date = "2022-02-01"

after_start_date = df_vac["date"] >= start_date
before_end_date = df_vac["date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = df_vac.loc[between_two_dates]
filtered_dates
```

```python
dates= pd.date_range('2021-02-01','2022-02-01' , freq='1M')-pd.offsets.MonthBegin(1)
dates
```

```python
mydf = pd.concat([filtered_dates.iloc[:,0:2],filtered_dates.iloc[:,11]],axis=1)
mydf.dropna(inplace = True)
```

```python
mydf = mydf.groupby(["location"]).sum().reset_index()
#mydf['year'] = pd.DatetimeIndex(mydf['date']).year
ind = mydf[~pd.DataFrame(mydf["location"].tolist()).isin(states["City"].tolist()).any(1).values].index
newdf = mydf.drop(ind)
cleaned_df = newdf.append(mydf[mydf["location"]=="New York State"])
cleaned_df["location"].replace({"New York State": "New York"}, inplace=True)
cleaned_df.sort_values(by=["location"]).reset_index(drop=True)["location"]
add_long = cleaned_df.merge(states.iloc[:,],left_on="location",right_on="City")
add_long.drop(['City'], axis=1,inplace = True)
final_df = add_long.reindex(columns = ["Longitude","Latitude","location","daily_vaccinations","State"])
final_df["ratio"]= final_df['daily_vaccinations']/final_df['daily_vaccinations'].sum()
final_df.drop(8).reset_index(drop = True,inplace = True)
len(final_df)
```

```python
import folium

# Make an empty map
us_map = folium.Map(location=[36,-100],zoom_start=4,tiles=None,overlay=False)
fg1 = folium.FeatureGroup(name='New Covid-19 Cases Past 7 Days',overlay=False)
fg2 = folium.FeatureGroup(name='Percent of Positive Cases Past 7 Days',overlay=False)
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = f'{url}/us-states.json'

folium.Choropleth(
    geo_data=state_geo,
    data=final_df,
    columns=["State", "ratio"],
    key_on="feature.id",
    fill_color="PuBuGn",
    fill_opacity= 0.7,
    line_opacity= 0.3,
    legend_name="United states vaccination rates (%)"
).add_to(fg1).add_to(us_map)
```

```python
for i in range(0,len(final_df)):
    
    popup_text = "{} people in {} has vaccinated\n, which is {:.2f} of total vaccinated population".format(
        final_df.iloc[i]["daily_vaccinations"],
        final_df.iloc[i]['location'],
        final_df.iloc[i]["ratio"])
    folium.Circle(
        location=[final_df.iloc[i]['Latitude'], final_df.iloc[i]['Longitude']],
        radius= final_df.iloc[i]['ratio']*5000000,
        color='crimson',
        fill=True,
        fill_color='crimson',
        popup = popup_text,
       
   ).add_to(fg2).add_to(us_map)
```

```python
folium.TileLayer('cartodbdark_matter',overlay=True,name="View in Dark Mode").add_to(us_map)
folium.TileLayer('cartodbpositron',overlay=True,name="Viw in Light Mode").add_to(us_map)
folium.LayerControl(collapsed=False).add_to(us_map)
us_map
```

```python
url3 = 'https://raw.githubusercontent.com/chengliu6/STA-141B-FINAL-PROJECT/main/full_data.csv'
df3 = pd.read_csv(url3)
df3["date"] = df3["date"].astype("datetime64")
df3
```

```python
after_start_date = df3["date"] >= start_date
before_end_date = df3["date"] <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = df3.loc[between_two_dates]
filtered_dates
df_cases = filtered_dates.iloc[:,0:4]
df_cases
```

```python
df_cases.dropna(inplace = True)
df_cases["new_deaths"].clip(lower=0, inplace = True)
df_cases["new_cases"].clip(lower=0, inplace = True)
```

```python
dindex = [x==df_cases["location"] for x in ["Asia",'Europe','European Union',
                                            'High income','Low income','Lower middle income',
                                            'North America','Oceania','South Africa','South America',"World"]]
```

```python
for i in dindex:
    df_cases.drop(index = df_cases[i].index, inplace = True)
```

```python
import calendar
import matplotlib.pyplot as plt

df_cases["new_cases"].plot()
```

```python
df_cases.set_index("date",inplace = True)
df_cases.plot(subplots=True, figsize=(10,10))
```

```python
df_cases["new_death_percentage"] = df_cases["new_deaths"].div(df_cases["new_cases"])
```

```python
df_cases["new_death_percentage"].plot()
```

```python
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px

df = df_cases.reset_index()

app = JupyterDash(__name__)


app.layout = html.Div([
    dcc.Dropdown(
                df['location'].unique(),
                value="Afghanistan",
                id='xaxis-column'
            ),
    dcc.RadioItems(
                ['new_cases', 'new_deaths',"new_death_percentage"],
                value = 'new_cases',
                id='xaxis-type',
                inline=True
            ),
    
    dcc.Graph(id='indicator-graphic'),

])


@app.callback(
    Output('indicator-graphic', 'figure'),
    Input('xaxis-column', 'value'),
    Input('xaxis-type', 'value'))
def update_graph(xaxis_column,
                 xaxis_type):
    
    fig = px.scatter(x=df[df["location"]==xaxis_column]["date"],y=df[df["location"]==xaxis_column][xaxis_type])
    
    return fig


app.run_server(debug=True)
```
