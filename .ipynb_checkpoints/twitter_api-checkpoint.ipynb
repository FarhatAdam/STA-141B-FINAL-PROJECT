```python
def get_covid_tweets(lat,long):
    tweet_list = []
    response = api.search_tweets("covid",
                                geocode = str(lat) + "," + str(long) + ",100km",
                                count = 10,
                                result_type="recent",
                                lang="en")
    for i in response: 
        tweet_list.append(i.text.split("https")[0])
    
    return tweet_list
```
